```python
#load census data and convert into a data frame
census_data = os.path.join('Data Sets',"acs2015_census_tract_data.csv")
census_df = pd.read_csv(census_data, skipinitialspace = True,low_memory = False)
census_df.head()
```

```python

m1_average=m1.groupby(['StateName','ReportYear'])[['Value']].mean()
#m1_average.head
#gradeReading = students_df1.groupby(['grade','school'])[['reading_score']].mean() 
#each_school_df["math_score"].groupby(each_school_df["school"]).mean()
```

```python
m1_df=pd.DataFrame(m1_average)
m1_df.head()
```

```python
measure_df = m1_df.reset_index()
measure_df.head()
```

```python
state_col = measure_df['StateName']
year_col = measure_df['ReportYear']

value_col = measure_df['Value']
value_plot= go.Bar(x = state_col, y = value_col, name = year_col,
                   marker=dict(color='rgb(158,202,225)'))
layout = dict(title = 'Air Quality Index Value(PM2.5) By State',
              xaxis = dict(title = 'State'),
              yaxis = dict(title = 'AQI Value'),
              )
data = [value_plot]
fig = dict(data=data,layout=layout)
py.iplot(fig)
```

```python
year2003_df = year2003.reset_index()
year2003_df.head()
```

```python
year2003_df = year2003_df.rename(columns={"StateName":"State"
                                            })
year2003_df.head()
```

```python
merged_data_df = pd.merge(year2003_df,average_income_df,on=['State'],how='outer')
merged_data_df.head()
```

```python
trace1 = go.Bar(
    x= merged_data_df['State'],
    y=merged_data_df['Income'],
    name='Income',
    marker=dict(color='rgb(158,202,225)')
)
trace2 = go.Scatter(
    x=merged_data_df['State'],
    y=merged_data_df['Value'],
    
    name='Ozone Concentration value',
    yaxis = 'y2'
)

data = [trace1, trace2]
layout = go.Layout(
    title='Income vs AQI(Ozone Concentration)',
    xaxis = dict(title='States'),
    yaxis=dict(
        title='Income'
    ),
    yaxis2=dict(
        title='AQI(ozone_concentration) value',
        titlefont=dict(
            color='rgb(153, 204, 255)'
        ),
        tickfont=dict(
            color='rgb(148, 103, 189)'
        ),
        overlaying='y',
        side='right'
    )
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```
