```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
census_data = os.path.join('Data Sets',"acs2015_census_tract_data.csv")
census_df = pd.read_csv(census_data, skipinitialspace = True,low_memory = False)
census_df.head()
```

```python
air_quality_data = os.path.join('Data Sets',"Air_Quality_data.csv")
air_quality_df = pd.read_csv(air_quality_data, skipinitialspace = True)
air_quality_df.head()
```

```python
renamed_df = air_quality_df.rename(columns={"StateName":"State",
                                            "CountyName":"County"})
renamed_df.head()
```

```python
merged_data_df = pd.merge(census_df,renamed_df,on=['County','State'],how='outer')
merged_data_df.head()
```

```python
merged_data1_df = merged_data_df[['CensusTract','County','State','Income','Value']]
merged_data1_df.head()
```
